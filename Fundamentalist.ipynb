Análise fundamentalista

Fatores determinantes da rentabilidade futura:

-Política de pagamento de dividendos da empresa;

-Desempenho da empresa e de sua posição no respectivo setor de atuação;

-Intensidade da concorrência e da existência de produtos e serviços alternativos;

-Nível de intervenção estatal na área de atuação da empresa;

-Programa de investimentos em bens de capital da empresa;

```python
newColumns = {
    'Nome': 'EMPRESA',
    'Classe': 'CLASSE',
    'Código': 'ATIVO',
    'PU\n26Fev20\najust p/ prov\nEm moeda orig': 'PRECO_ACAO',
    'Qtd Ações\nOutstanding\nda empresa\nem milhares\n26Fev20': 'QTD_ACOES',
    'LPA\n Mais recente\n Em moeda orig\n de 12 meses\n consolid:sim*\n ajust p/ prov': 'LUCRO_P_ACAO',
    'Patrim Liq\n Mais recente\n Em moeda orig\n em milhares\n consolid:sim*': 'PATRIMONIO_LIQ',
    'EV\n26Fev20\nEm moeda orig\nem milhares\nconsolid:sim*': 'ENTERPRISE_VALUE',
    'EBITDA\n Mais recente\n Em moeda orig\n em milhares\n de 12 meses\n consolid:sim*': 'EBITDA',
    'Div Tt Lq\n Mais recente\n Em moeda orig\n em milhares\n consolid:sim*': 'DIV_LIQ',
    'Lucro Liquido\n Mais recente\n Em moeda orig\n em milhares\n no exercício\n consolid:sim*': 'LUCRO_LIQ',
    'DivPag\n Mais recente\n Em moeda orig\n em milhares\n no exercício\n consolid:sim*': 'DIVIDENDOS',
    'Payout Ratio': 'PAYOUT_RATIO',
    'Div Yld (inic)\n26Fev20\n1 anos\nEm moeda orig': 'DIV_YIELD',
    'VPA': 'VPA',
    'P/VPA': 'PVPA',
    'EV/EBITDA': 'EVEBITDA',
    'P/L': 'PL',
    'ROIC (IC fim) %\n Mais recente\n de 12 meses\n consolid:sim*': 'ROIC',
    'RenPat(fim)\n Mais recente\n de 12 meses\n consolid:sim*': 'ROE',
    'LiqSec\n Mais recente\n consolid:sim*': 'LIQ_SECA',
    'LiqCor\n Mais recente\n consolid:sim*': 'LIQ_COR',
    'GiroAt\n Mais recente\n de 12 meses\n consolid:sim*': 'GIRO_ATIVO',
    'Alavancagem': 'ALAVANCAGEM',
    'Margem Liquida\n Mais recente\n de 12 meses\n consolid:sim*' : 'MARG_LIQ'
}
df.rename(newColumns,axis=1,inplace=True)
colunasNumericas = [
    'PRECO_ACAO',
    'QTD_ACOES',
    'LUCRO_P_ACAO',
    'PATRIMONIO_LIQ',
    'ENTERPRISE_VALUE',
    'EBITDA',
    'DIV_LIQ',
    'LUCRO_LIQ',
    'DIVIDENDOS',
    'PAYOUT_RATIO',
    'DIV_YIELD',
    'VPA',
    'PVPA',
    'EVEBITDA',
    'PL',
    'ROIC',
    'ROE',
    'LIQ_SECA',
    'LIQ_COR',
    'GIRO_ATIVO',
    'ALAVANCAGEM',
    'MARG_LIQ'
]
df.replace('-', 0, inplace=True)
df[colunasNumericas] = df[colunasNumericas].astype(float)
df.describe()
```

```python
newColumns = [
    'EMPRESA',
    'ATIVO',
    'PVPA',
    'PL',
    'ROE',
    'LIQ_COR',
    'ALAVANCAGEM',
    'MARG_LIQ',
    'DIV_LIQ',
    'LUCRO_LIQ',
    'PAYOUT_RATIO',
    'DIV_YIELD',
    'EVEBITDA'
]
df = df[newColumns]
```

INDICADORES:

PVPA: Valor de mercado / Valor patrimonio (Maior pior, mais alavancado)

PL: Payback (maior "pior")

ROE: Lucro liq / Patr liq (maior melhor)

LIQ_COR: Quanto tem agora de caixa/Quanto deve agora (maior melhor)

ALAVANCAGEM: Divida liquida / EBITDA  (menor melhor)

MARG_LIQ: O quanto resta depois de todas as operaçoes e pagamentos (maior melhor)

EVEBITDA: Quanto demoraria para a empresa se recomprar

```python
outliers = ["MOAR3"]

col = list(df.columns)
col.remove("EMPRESA")
col.remove("ATIVO")
df = df[(df.ROE > 8) & (df.ROE < 60)]
df = df[df.MARG_LIQ > 4.5]
df = df[df.LUCRO_LIQ > 0]
df = df[df.PL > 0]
df = df[df.PVPA < 12]
df = df[~df.ATIVO.isin(outliers)]
df.shape
```

```python


df.sort_values("DIV_YIELD")
df
```

```python
df1 = pd.DataFrame()
df["ALAVANCAGEM_deslocada"] = df.apply(lambda row: row.ALAVANCAGEM - df.ALAVANCAGEM.min(), axis=1)
df1 = df[["EMPRESA", "ATIVO"]].copy()
colunas = ["PVPA", "PL", "ROE", "LIQ_COR", "ALAVANCAGEM_deslocada", "MARG_LIQ", "DIV_YIELD"]
coeficientes_modo = [0, 1, 0, 0, 1, 0, 0]
coeficientes_peso = [0.5, -0.5, 2, 1, -1, 2, 3]
for coluna, modo, peso in zip(colunas, coeficientes_modo, coeficientes_peso):
    df1[coluna] = ((df[coluna] - df[coluna].min()) / (df[coluna].max() - df[coluna].min()) - modo) * peso

                   
df["ALAVANCAGEM_deslocada"].min()
df1["NOTA"] = df1[colunas].apply(lambda row: stats.hmean(row), axis=1)
df1 = df1.sort_values("NOTA", ascending=False)
```

```python
df.to_excel("Fund_Filtrado.xlsx", index=False)
df1.to_excel("Notas.xlsx", index=False)
```
