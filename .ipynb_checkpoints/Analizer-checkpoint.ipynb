```python
import pandas as pd
import seaborn as sb
from os import chdir, getcwd

initPath = getcwd()

chdir("./data/")

df = pd.read_csv("petr4.csv")
df.head()

chdir(initPath)
```

```python
#pd.plotting.scatter_matrix(df, figsize = (8,8))
df.describe()
```

```python
sb.regplot(data = df, x = 'open', y = 'close')
```

```python
df["variacao"] = df.apply(lambda row: abs(row.open - row.close), axis=1)
df["variacaoIntra"] = df.apply(lambda row: abs(row.high - row.low), axis=1)
df.describe()
```

```python
print(df.sort_values(by="variacao", ascending=False))
```

```python
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.relplot(x="timestamp", y="close", kind="line", data = df)
```
