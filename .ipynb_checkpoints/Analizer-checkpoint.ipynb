```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

df = pd.read_csv("final_data.csv")

#df = df.sort_values(["timestamp","ativo"])
#df = df[["timestamp", "ativo", "variacaoPercentual","open", "high", "low", "close", "volume", "variacaoDiaria", "variacaoIntra", "year"]]
```

```python
df = df.sort_values(["timestamp", "ativo"])
datas = list(set(df["timestamp"]))
datas.sort()
acoes = []
for dia in datas:
    df_dia = df[df["timestamp"] == dia]
    acoes.append(df_dia["ativo"].count())
    
acoes
```

```python
sb.lineplot( x=datas, y=acoes)
```

```python
#Obter as informaçoes de hoje
df = df[df["timestamp"] == "2020-01-30"]
df.groupby(["ativo"])
```

```python
df_preco = df[["ativo", "close"]]
df_preco = df_preco.sort_values("close")
df_preco = df_preco.groupby("ativo").mean()

df_preco
```

```python
df_preco.to_csv("final_data_analise_preco.csv")
```

```python
df_volume = df[["ativo", "volume"]]
df_volume = df_volume.groupby("ativo").mean()
df_volume = df_volume.sort_values("volume", ascending=False)
df_volume1 = pd.DataFrame()
for i in range(202):
    df_volume1 = df_volume1.append(df_volume.iloc[i])

index = df_volume1.index
index = list(index)
df_index = pd.DataFrame(index)
df_index.to_json("index.json", orient="values")
index.sort()

index
```

```python
df = df[df["ativo"].isin(index)]
df.to_csv("final_data_revised.csv", index=False)
```

```python
#Achar variacao percentual de todos os ativos

df_vp =  df.groupby('ativo')["variacaoPercentual"].mean()
df_vp.to_csv("variacaoPercentual.csv", header=["variacaoPercentual"])
df_vp.shape
```

```python

df_vp = df_vp.sort_values(ascending=False)
df_vp.to_csv("variacaoPercentual.csv", header=["variacaoPercentual"])
```

```python
#Análise SBSP3
sbsp = df[df["ativo"] == "VVAR3"]


sbsp.describe()
```

```python
indexClean = ["variacaoPercentual", "open", "close"]
for ind in indexClean:
    Q1 = sbsp.describe()[ind].iloc[4]
    Q3 = sbsp.describe()[ind].iloc[6]
    IQR = Q3 - Q1
    sbsp = sbsp[sbsp[ind] < Q3 + 2 * IQR]
    sbsp = sbsp[Q1 - 2 * IQR < sbsp[ind]]
sbsp.describe()
sbsp.shape
```

```python
close = sbsp["low"]

year = [x for x in range(sbsp.shape[0])]

sb.lineplot(x=year, y=close)
sb_plot = sb.lineplot(x=year, y=close)
plt.savefig("output.png", dpi=260)
```
